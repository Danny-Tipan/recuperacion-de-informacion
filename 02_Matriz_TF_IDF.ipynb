```python
import unicodedata
import re

ruta = "/content/01_corpus_turismo.txt"
with open(ruta, 'r', encoding='utf-8') as f:
    lineas = f.readlines()

def limpiar_texto(texto):
    texto_sin_tildes = ''.join((c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn'))
    texto_limpio = re.sub(r'[^\w\s]', '', texto_sin_tildes).lower()
    return texto_limpio

documentos = [limpiar_texto(linea) for linea in lineas]

matriz = {}

for i, doc in enumerate(documentos):
    doc_id = f'Doc{i+1}'
    palabras = doc.split()
    for palabra in palabras:
        if palabra not in matriz:
            matriz[palabra] = {}
        if doc_id not in matriz[palabra]:
            matriz[palabra][doc_id] = 0
        matriz[palabra][doc_id] += 1

documentos_ids = [f'Doc{i+1}' for i in range(len(documentos))]
terminos_ordenados = sorted(matriz.keys())

print("Término\t\t" + "\t".join(documentos_ids))
for termino in terminos_ordenados:
    fila = [str(matriz[termino].get(doc_id, 0)) for doc_id in documentos_ids]
    print(f"{termino:15}\t" + "\t".join(fila))
```
